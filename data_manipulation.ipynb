```python
import pandas as pd
import matplotlib.pyplot as plt
import pylab

col_names = ["Input", "MarDRe", "NGSReadsTreatment", "ParDRe", "FastUniq", "BioSeqZip"]
df = pd.read_csv('dados.csv', names=col_names)
#print(df)
df2 = df.sort_values(by='Input', ascending=True)
#print(df2)
```

```python
import numpy as np
from scipy.interpolate import *
import csv

input1 = [int(val) for val in df2.Input]
mar = [float(val.replace("[","").replace("]","")) for val in df2.MarDRe]
NGS = [float(val.replace("[","").replace("]","")) for val in df2.NGSReadTreatment]
par = [float(val.replace("[","").replace("]","")) for val in df2.ParDRe]
fast = [float(val.replace("[","").replace("]","")) for val in df2.FastUniq]
bioseq = [float(val.replace("[","").replace("]","")) for val in df2.BioSeqZip]
```

```python
plt.title('Algorithm Complexity')
plt.xlabel('Input Size')
plt.ylabel('Time Complexity')

plt.plot(input1, mar, label='MarDre')
plt.scatter(input1, mar, color='blue')

plt.plot(input1, NGS, label='NGSReadsTreatment')
plt.scatter(input1, NGS, color='orange')

plt.plot(input1, par, label='ParDre')
plt.scatter(input1, par, color='green')

plt.plot(input1, fast, label='FastUniq')
plt.scatter(input1, fast, color='red')

plt.plot(input1, bioseq, label='BioSeqZip')
plt.scatter(input1, bioseq, color='pink')

plt.grid()
plt.legend()
pylab.rcParams['figure.figsize'] = (8.0, 8.0)
plt.show()
#plt.savefig('Result.png')
```
